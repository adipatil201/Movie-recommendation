```python
#so, we need to convert:
#list of dict : [{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}] 
#to 
#list : ['Adventure','Fantasy','Action']
#for this,will define a helper function


def convert(obj):        #input obj will be a list of dict
    
    output=[]
    for i in obj:
        output.append(i['name'])
    
    return output        #output will be a list having all genres
```

```python
#but another problem is that genres stores all its data as a string
#so, first we need to convert strings from genre to a list
#this can be achieved using the ast module of python

import ast
ast.literal_eval('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

```python
# the correct convert func will be

def convert(obj):       
    
    output=[]
    for i in ast.literal_eval(obj):
        output.append(i['name'])
    
    return output   
```

```python
convert('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

```python
#for this cast column, we are only interested in the name of the actors from {"cast_id": 4, "character": "Captain Jack Sparrow", "credit_id": "52fe4232c3a36847f800b50d", "gender": 2, "id": 85, "name": "Johnny Depp", "order": 0}
#as all other data is irrelevant. this name is stored in the 'name' for each dictionary
#also, not all actors are inportant, so we will use the top 10 actors from each movie

def convert2(obj):       
    c=0
    output=[]
    
    for i in ast.literal_eval(obj):
        if c<=10:
            output.append(i['name'])
            c+=1
        else:
            break
        
    return output   
```

```python
#now, we will convert these tags into a meaningful vector representation so that we can use them to make find similar movies
#one way to convert all the tags into a vector rep is to
#i) add all the tags to form a large corpus of text
#ii) from this corpus, find the top 5000 words which appear in this corpus
#iii) check how many times each of the 5000 most occuring words have occured in the tags of a given movie
#this gives us a 5000-dim vector for each of the movies which represents the tags 
#also, while checking the 5000 most occuring words, we will remove all the words like the,in,and,of,etc i.e stop words will be removed
#also, we only use the root of a word i.e words like run,running,ran,etc will only be represented by a single root word; this is also called stemmimg
```

```python
#the entire task mentioned above can directly be done using the Scikit learn library

from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english')       #this an obj which will be used coverts the large text corpus into vectors; max_features tells the max dimensions of vector to be made and stop_words=engllish removes all the stop words of english language while converting the text to vector
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()               # this is the porter stemmer obj which will be used to stem the tags of ecah movie before vectorizing them
```

```python
#eg

l=['love','loving','loved']

for i in range(0,len(l)):
    l[i]=ps.stem(l[i])
    
print(l)
```

```python
def stem_transformer(obj):          # input obj will be a string of tags
    l=obj.split()             #converts string to list of words
    for i in range(0,len(l)):
        l[i]=ps.stem(l[i])        # performs stemming on each word of the list
    k=" ".join(l)                 #joins all words of list at every space
    return k
```

```python
#now, based on only this similarity scores, we will make a recommender function

def basic_recommender(movie_name,movie_database=new_df,similarity_score=sim):
    movie_id=movie_database.id[movie_database['title']==movie_name]            #gets the movie id of the entered movie name
    movie_index=movie_database[movie_database['title']==movie_name].index      #gets the list index positions of the movie using its title in the new_df
    
    l=list(enumerate(similarity_score[movie_index[0]]))                        # l is a list of tuples such that l[i]=(i,similarity_score[movie_index[0]])
                                                                               #this was done to preserve the index positions of all the similarity_score
        
    l=sorted(l,reverse=True,key=lambda x:x[1])                                 #sorts the list l in descending order based on the 2nd value of each tuple in the list
    
    best_rcmds=l[1:11]                                                          #top 5 most similar movies to the current movie
    
    index_of_best_rcmds=[]                                                     
    for i in best_rcmds:                                                       #used to get indices of the best recomendations
        index_of_best_rcmds.append(i[0])                                       
    
    titles_of_best_rcmds=[]                                                    #used to store titles of best recommendations
    for i in index_of_best_rcmds:
        titles_of_best_rcmds.append(movie_database.title[i])
        
    ids_of_best_rcmds=[]                                                       #used to store ids of best recommendations
    for i in index_of_best_rcmds:
        ids_of_best_rcmds.append(movie_database.id[i])
        
    return titles_of_best_rcmds                                              #returns names of top 5 recommendations
```

```python
from keras.preprocessing.text import text_to_word_sequence
```

```python
def t2v(obj,dict_of_word_embeddings=embeddings_index):
    
    final_embedding_of_text=np.zeros((300,))
    c=0
    
    words_list = text_to_word_sequence(obj)
    for i in words_list:
        if i in embeddings_index:
            final_embedding_of_text=final_embedding_of_text+embeddings_index[i]
            c=c+1
        else:
            continue
    
    final_embedding_of_text=final_embedding_of_text/c
    
    return final_embedding_of_text
    
```

```python
def t2v2(obj,model=model):
    
    final_embedding_of_text=np.zeros((300,))
    c=0
    
    words_list = text_to_word_sequence(obj)
    for i in words_list:
        try:
            c+=1
            final_embedding_of_text = final_embedding_of_text + model.wv[i]
        except KeyError:
            continue
            
    final_embedding_of_text=final_embedding_of_text/c
    
    return final_embedding_of_text
    
```
